1. Write a Python program to create a class representing a Circle. Include methods to calculate its area and perimeter.

```python

import math  

class Circle:
    def __init__(self, radius):
        self.radius = radius  

    # Method to calculate area
    def area(self):
        return math.pi * self.radius ** 2

    # Method to calculate perimeter (circumference)
    def perimeter(self):
        return 2 * math.pi * self.radius


r = float(input("Enter the radius of the circle: "))
c = Circle(r)

print(f"Area of the circle: {c.area():.2f}")
print(f"Perimeter (Circumference) of the circle: {c.perimeter():.2f}")
```

2. Write a Python program to create a person class. Include attributes like name, country and date of birth. Implement a method to determine the person's age.


```python
from datetime import date

class Person:
    def __init__(self, name, country, year, month, day):
        self.name = name
        self.country = country
        self.year = year
        self.month = month
        self.day = day

    # Method to calculate age
    def age(self):
        today = date.today()
        age = today.year - self.year
        if (today.month, today.day) < (self.month, self.day):
            age -= 1
        return age

name = input("Enter name: ")
country = input("Enter country: ")
year = int(input("Enter birth year (YYYY): "))
month = int(input("Enter birth month (1-12): "))
day = int(input("Enter birth day (1-31): "))

person = Person(name, country, year, month, day)

print(f"\nName: {person.name}")
print(f"Country: {person.country}")
print(f"Age: {person.age()} years")
```

3. Write a Python program to create a calculator class. Include methods for basic arithmetic operations.

```python

class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    # Method for addition
    def add(self):
        return self.a + self.b

    # Method for subtraction
    def subtract(self):
        return self.a - self.b

    # Method for multiplication
    def multiply(self):
        return self.a * self.b

    # Method for division
    def divide(self):
        if self.b != 0:
            return self.a / self.b
        else:
            return "Error! Division by zero."

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

calc = Calculator(num1, num2)

print(f"\nAddition: {calc.add()}")
print(f"Subtraction: {calc.subtract()}")
print(f"Multiplication: {calc.multiply()}")
print(f"Division: {calc.divide()}")
```
